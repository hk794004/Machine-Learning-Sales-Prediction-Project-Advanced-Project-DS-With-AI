```python
# Import Libraries

import pandas as pd
import numpy as np
import streamlit as st
import matplotlib as plt
import seaborn as sns
import joblib
import pickle

from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, OrdinalEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression

def Load_Data(Csv_Path):
    return pd.read_csv(Csv_Path)

Data = Load_Data('Sales_.csv')

Data = Data.drop(['Unnamed: 25','Unnamed: 26','Unnamed: 27','Expense','% Discount','discount','Packaging Cost','Service and other cost','Total Cost','Profit','Item Price','Order_ID','Ship Date','Ship Mode','Currency','Product Name','customer name'], axis= 1)

Data['Order Date'] = pd.to_datetime(Data['Order Date'])

Data['Order_Day'] = Data['Order Date'].dt.day_name()

Data['Order_Month'] = Data['Order Date'].dt.month_name()

Data['Order_Year'] = Data['Order Date'].dt.year

Data['Week_No'] = Data['Order Date'].dt.weekday

X = Data.drop(['Order Date','Sales'],axis = 1)

Y = Data['Sales']

Cat_Cols = [cat for cat in ['Country','City','State','Region','Category','Sub_Category','Order_Day','Order_Month'] if cat in X.columns]

Num_Cols = [num for num in X.columns if num not in Cat_Cols]

Categorical_Transformer = Pipeline(
    steps=[
        ('imputer',SimpleImputer(strategy='most_frequent')),
        ('one_hot',OneHotEncoder(handle_unknown='ignore'))
        ]
)

Numerical_Transformer = Pipeline(
        steps=[
            ('imputer',SimpleImputer(strategy='median')),
            ('scaler',StandardScaler())
        ]
)

Preprocessor = ColumnTransformer(
        transformers=[
            ('cat',Categorical_Transformer,Cat_Cols),
            ('num',Numerical_Transformer,Num_Cols),
        ]
)

Model = Pipeline(
        steps=[
            ('preprocessor',Preprocessor),
            ('regression',LinearRegression())
        ]
)

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2,random_state=42)

Model.fit(X_train,Y_train)

Y_Pred = Model.predict(X_test)

Test_Metrics = {
        'mae' : mean_absolute_error(Y_test,Y_Pred),
        'mse' : mean_squared_error(Y_test,Y_Pred),
        'r2' : r2_score(Y_test,Y_Pred),
}

Pred_Y = Model.predict(X_train)

Train_Metrics = {
        'mae' : mean_absolute_error(Y_train,Pred_Y),
        'mse' : mean_squared_error(Y_train,Pred_Y),
        'r2' : r2_score(Y_train,Pred_Y),
}
```

```python
joblib.dump(Model,'Model_Pipeline.pkl')

joblib.dump(Train_Metrics,'Train_Metrics.pkl')

joblib.dump(Test_Metrics,'Test_Metrics.pkl')
```
